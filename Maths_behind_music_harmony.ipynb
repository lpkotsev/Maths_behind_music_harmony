### 2. Introduction

Music is fundamentally a physical and mathematical phenomenon. Every musical note corresponds to a sound wave vibrating at a specific frequency. When two or more notes are played together, their interaction creates what humans perceive as harmony.

Understanding why certain combinations of notes sound pleasant while others do not has been a central question in both music theory and mathematics.

One of the earliest mathematical investigations of music is attributed to the ancient Greek mathematician Pythagoras (Rossing, 2002). Through experiments with vibrating strings, he discovered that harmonious musical intervals correspond to simple ratios between frequencies.

For example:

Octave → ratio 2:1  
(meaning that if a note vibrates at e.g. 440 Hz, then the octave higher will vibrate twice more at 880 Hz)
Perfect fifth → ratio 3:2  
(meaning that if a note vibrates at e.g. 440 Hz, then the perfect fifth will vibrate 1.5 more at 660 Hz)
Perfect fourth → ratio 4:3  
(meaning that if a note vibrates at e.g. 440 Hz, then the perfect fifth will vibrate 1.5 more at 587 Hz)

When the waves align in this way, the combination of notes sounds stable and pleasant to our ears. This is why these intervals are commonly used in music and are considered harmonious. In other words, these simple ratios form the mathematical foundation of musical harmony (Benson, 2007; Hall, 2008). 

The main objective of this project is to explore how mathematical relationships between frequencies explain musical harmony and how these relationships can be modeled and simulated using Python.

Understanding these relationships is important not only for music theory but also for modern applications such as digital sound synthesis, music production, acoustic engineering, and audio signal processing.

### 5.1 Mathematical Model of Musical Notes

Every musical note corresponds to a sound wave vibrating at a particular frequency. Frequency is measured in Hertz (Hz) and represents the number of oscillations per second.

In modern music theory, the reference pitch is defined as the note A4 with a frequency of:

A4 = 440 Hz

Musical notes are organized into octaves. When moving one octave higher, the frequency doubles.
This doubling relationship forms the basis of musical pitch perception.

Mathematically:

$$
f_{\text{octave}} = 2f
$$


This mathematical formula allows us to generate the entire chromatic scale programmatically.
### 5.2 Equal Temperament Scale

Modern Western music uses the 12-tone equal temperament tuning system. In this system, an octave is divided into twelve equal steps called semitones. This system allows instruments to play in different keys while maintaining consistent tuning relationships (Benson, 2007).

Each semitone multiplies the frequency by the constant ratio:

$$
2^{1/12}
$$

If a base frequency is denoted as f, the frequency of a note n semitones above it can be calculated using:

$$
f_n = f \cdot 2^{n/12}
$$

This exponential relationship ensures that after twelve steps the frequency doubles, completing the octave.


To better understand the equal temperament system, we visualize the frequencies of the chromatic scale.

The resulting diagram shows how frequencies increase as we move from one note to the next. Because each step multiplies the frequency by the same ratio, the growth follows an exponential pattern.

```python
plt.show()
import numpy as np
import matplotlib.pyplot as plt

A4 = 440
semitone_ratio = 2 ** (1/12)

notes = [
"C","C#","D","D#","E","F",
"F#","G","G#","A","A#","B"
]

frequencies = []

for i in range(-9,3):
    freq = A4 * (semitone_ratio ** i)
    frequencies.append(freq)

plt.figure(figsize=(10,5))

plt.plot(notes, frequencies, marker='o')

plt.title("Chromatic Scale Frequencies")
plt.xlabel("Musical Notes")
plt.ylabel("Frequency (Hz)")
plt.grid(True)
```

```python
This visualization confirms that the structure of modern musical tuning is fundamentally exponential.
The plot helps illustrate several important properties:

• frequencies increase continuously across the scale  
• the distance between notes is multiplicative rather than additive  
• equal temperament produces evenly spaced musical steps on a logarithmic scale
```

The diagram therefore illustrates the mathematical basis of musical consonance.

### 5.4 Harmonic Series

When a musical instrument produces sound, it does not generate a single frequency. Instead, it produces a fundamental frequency together with several higher frequencies called harmonics or overtones.

These harmonics follow the harmonic series:

$f_n = n \cdot f$

where n is an integer and f is the fundamental frequency.

For example, if the fundamental frequency is 440 Hz, the harmonics will be:

440 Hz  
880 Hz  
1320 Hz  
1760 Hz

The harmonic series explains why certain musical intervals appear naturally in music. Many important intervals such as octaves and fifths occur directly within this sequence.  These relationships help explain why certain intervals sound consonant to the human ear (Hall, 2008).


```python
plt.show()
import numpy as np
import matplotlib.pyplot as plt

base_freq = 440

harmonic_numbers = np.arange(1,9)
harmonics = base_freq * harmonic_numbers

plt.figure(figsize=(10,5))

plt.stem(harmonic_numbers, harmonics)

plt.xlabel("Harmonic Number")
plt.ylabel("Frequency (Hz)")
plt.title("Harmonic Series and Musical Intervals")

# Label important intervals
plt.text(2, harmonics[1]+50, "Octave\n2:1", ha='center')
plt.text(3, harmonics[2]+50, "Perfect Fifth\n3:2", ha='center')
plt.text(4, harmonics[3]+50, "Perfect Fourth\n4:3", ha='center')

plt.grid(True)
```

The diagram shows the harmonic series generated from a fundamental frequency of 440 Hz.

Each vertical line in the diagram represents one harmonic. The horizontal axis shows the harmonic number (1st harmonic, 2nd harmonic, etc.), while the vertical axis shows the corresponding frequency in Hertz.

The first harmonic corresponds to the fundamental frequency itself (440 Hz). Each subsequent harmonic is an integer multiple of this base frequency. For example, the second harmonic has frequency 880 Hz, the third harmonic 1320 Hz, and the fourth harmonic 1760 Hz. The third harmonic is Perfect Fifth because the division of it and the previous harmonic would give the ratio 3:2. We could find other Perfect Fifths from lower octaves by finding the lower octave from the third harmonic 1320 Hz / 2 = 660 Hz. Now 660 Hz would be a perfect fifth to 440Hz as well because 660 Hz / 440 Hz would give the 3:2 ratio (equivalent to perfect fifth)



This pattern demonstrates that the frequencies produced by a vibrating object follow a regular mathematical structure. Because these frequencies are related by simple integer multiples, many important musical intervals naturally appear in the harmonic series. For example, the second harmonic forms an octave above the fundamental frequency, and the third harmonic forms an interval close to a perfect fifth.

The diagram therefore illustrates how musical harmony arises naturally from the physics of vibrating sound sources.

### 5.5 Computational Sound Generation


To demonstrate the mathematical structure of musical harmony in practice, this project generates sound waves directly using Python.

In digital audio synthesis, musical tones are commonly modeled as sinusoidal waves. This approach is widely used in computer music and digital sound processing because sine waves represent the simplest form of a periodic sound signal (Roads, 1996).

A pure musical tone can be represented mathematically using the function:

y(t) = sin(2πft)

where:

f = frequency of the sound (measured in Hertz)  
t = time variable




This equation describes a sound wave that oscillates smoothly over time. The frequency f determines how fast the wave oscillates, which corresponds to the pitch of the sound that we hear.

For example, if the frequency is 440 Hz, the wave completes 440 oscillations per second, producing the musical note A4.

Using Python, this equation can be used to generate digital sound signals. By computing the values of this function over small time intervals, we obtain a sequence of numbers that represents the sound wave. This sequence can then be played as audio.


```python
import numpy as np
import matplotlib.pyplot as plt

# frequency of the tone
f = 440  # A4 note

# time axis
t = np.linspace(0, 0.01, 1000)

# sine wave equation
y = np.sin(2 * np.pi * f * t)

plt.figure(figsize=(10,4))

plt.plot(t, y)

plt.title("Sine Wave Representation of a Musical Tone")
plt.xlabel("Time (seconds)")
plt.ylabel("Amplitude")

plt.grid(True)

plt.show()
```

### 5.6 Circle of Fifths

Another important mathematical structure used in music theory is the circle of fifths. 
The circle of fifths organizes musical notes according to repeated applications of the perfect fifth interval.

A perfect fifth corresponds to the frequency ratio:

3 : 2

If a note with frequency f is multiplied by this ratio, the resulting note forms a perfect fifth above the original note.

f_new = (3/2)f

By repeatedly applying this relationship, a sequence of notes is obtained:

C → G → D → A → E → B → F#

This sequence forms the circle of fifths, which plays a central role in tonal harmony and chord progressions. 
Many common harmonic progressions move between chords that are neighbors in the circle of fifths.

The circle of fifths therefore provides a mathematical framework for understanding how harmonic relationships are structured in music.

## Appendix A: Generating Musical Chords

This appendix presents the Python implementation used to generate and play musical chords. 
The code demonstrates how mathematical frequency ratios can be translated into audible sound.

A musical tone can be represented as a sine wave. By generating sine waves for several frequencies and combining them, we can simulate musical chords.

For example, a major chord consists of three notes:

• Root note  
• Major third  
• Perfect fifth  

These correspond to the frequency ratios:

1 : 5/4 : 3/2

If the root frequency is 440 Hz, the frequencies of the chord are:

440 Hz (root)  
550 Hz (major third)  
660 Hz (perfect fifth)

The following code generates and plays this chord.

```python
import numpy as np
from IPython.display import Audio

def generate_tone(frequency, duration=1.5, sample_rate=44100):

    t = np.linspace(0, duration, int(sample_rate * duration), False)
    tone = np.sin(2 * np.pi * frequency * t)

    return tone

def major_chord(root_freq, duration=2):

    root = generate_tone(root_freq, duration)
    third = generate_tone(root_freq * 5/4, duration)
    fifth = generate_tone(root_freq * 3/2, duration)

    chord = root + third + fifth

    chord = chord / np.max(np.abs(chord))

    return chord

Audio(major_chord(440), rate=44100)
```

## Appendix B: Generating Musical Scales

This appendix presents the Python implementation used to generate and play a musical scale.

Modern Western music typically uses the **equal temperament tuning system**, where the octave is divided into twelve equal semitone steps.

Each semitone increases the frequency by the factor:

2^(1/12)

Using this mathematical relationship, we can compute the frequencies of notes in a musical scale and generate the corresponding sound waves.

```python
import numpy as np
from IPython.display import display, Audio

def generate_tone(freq, duration=0.4, samplerate=44100):
    t = np.linspace(0, duration, int(samplerate * duration), endpoint=False)
    tone = 0.5 * np.sin(2 * np.pi * freq * t)

    fade_len = int(0.01 * samplerate)
    envelope = np.ones_like(tone)
    envelope[:fade_len] *= np.linspace(0, 1, fade_len)
    envelope[-fade_len:] *= np.linspace(1, 0, fade_len)

    return tone * envelope

def play_scale(frequencies):
    parts = [generate_tone(f) for f in frequencies]
    melody = np.concatenate(parts)
    melody = melody / np.max(np.abs(melody))
    return melody.astype(np.float32)

melody = play_scale([261.63, 293.66, 329.63, 349.23, 392.00])

display(Audio(melody, rate=44100))
```
