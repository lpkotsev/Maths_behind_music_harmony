When two musical notes sound pleasant together (a consonance), their frequencies often form simple ratios such as 2:1 or 3:2. These ratios form the mathematical foundation of musical scales, chords, and harmony.

The goal of this project is to explore the mathematical relationship between frequency ratios and musical harmony,
how musical scales are constructed mathematically and how harmony can be simulated and analyzed using Python
