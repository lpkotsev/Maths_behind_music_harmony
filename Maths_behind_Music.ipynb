## 2. Introduction

Music is fundamentally a physical and mathematical phenomenon. Every musical note corresponds to a sound wave vibrating at a specific frequency. When two or more notes are played together, their interaction creates what humans perceive as harmony.

Understanding why certain combinations of notes sound pleasant while others do not has been a central question in both music theory and mathematics.

One of the earliest mathematical investigations of music is attributed to the ancient Greek mathematician Pythagoras (Rossing, 2002). Through experiments with vibrating strings, he discovered that harmonious musical intervals correspond to simple ratios between frequencies.

For example:

Octave → ratio 2:1  
(meaning that if a note vibrates at e.g. 440 Hz, then the octave higher will vibrate twice more at 880 Hz)
Perfect fifth → ratio 3:2  
(meaning that if a note vibrates at e.g. 440 Hz, then the perfect fifth will vibrate 1.5 more at 660 Hz)
Perfect fourth → ratio 4:3  
(meaning that if a note vibrates at e.g. 440 Hz, then the perfect fifth will vibrate 1.5 more at 587 Hz)

When the waves align in this way, the combination of notes sounds stable and pleasant to our ears. This is why these intervals are commonly used in music and are considered harmonious. In other words, these simple ratios form the mathematical foundation of musical harmony (Benson, 2007; Hall, 2008).

### 3.1 Equal Temperament Scale

Modern Western music uses the 12-tone equal temperament tuning system (Benson, 2007). In this system, an octave is divided into twelve equal steps called semitones. This system allows instruments to play in different keys while maintaining consistent tuning relationships (Benson, 2007).

Each semitone multiplies the frequency by the constant ratio (Benson, 2007):

$$
2^{1/12}
$$

If a base frequency is denoted as f, the frequency of a note n semitones above it can be calculated using:

$$
f_n = f \cdot 2^{n/12}
$$

This exponential relationship ensures that after twelve steps the frequency doubles, completing the octave (Benson, 2007).


```python

import numpy as np
import matplotlib.pyplot as plt

A4 = 440
semitone_ratio = 2 ** (1/12)

notes = [
"C","C#","D","D#","E","F",
"F#","G","G#","A","A#","B"
]

frequencies = []

for i in range(-9,3):
    freq = A4 * (semitone_ratio ** i)
    frequencies.append(freq)

plt.figure(figsize=(10,5))

plt.plot(notes, frequencies, marker='o')

plt.title("Chromatic Scale Frequencies")
plt.xlabel("Musical Notes")
plt.ylabel("Frequency (Hz)")
plt.grid(True)
```

• the distance between notes is multiplicative rather than additive  
• equal temperament produces evenly spaced musical steps on a logarithmic scale

### 3.2 Frequency and Pitch Relationships in Musical Notes

Every musical note corresponds to a sound wave vibrating at a particular frequency. Frequency is measured in Hertz (Hz) and represents the number of oscillations per second.

Musical notes are organized into octaves. As mentioned in the above subpoint about equal temperament scale when moving one octave higher, the frequency doubles (Rossing, 2002).
This doubling relationship forms the basis of musical pitch perception.

Mathematically:

$$
f_{\text{octave}} = 2f

## 3.3 Harmonic Series

When a musical instrument produces sound, it does not generate a single frequency. Instead, it produces a fundamental frequency together with several higher frequencies called harmonics or overtones (Hall, 2008; Rossing, 2002).

These harmonics follow the harmonic series:

$f_n = n \cdot f$  (Hall, 2008)

where n is an integer and f is the fundamental frequency.

For example, if the fundamental frequency is 440 Hz, the harmonics will be:

440 Hz  
880 Hz  
1320 Hz  
1760 Hz

The harmonic series explains why certain musical intervals appear naturally in music.  These relationships help explain why certain intervals sound consonant to the human ear (Hall, 2008). This pattern also demonstrates that the frequencies produced by a vibrating object follow a regular mathematical structure. Because these frequencies are related by simple integer multiples, many important musical intervals naturally appear in the harmonic series. For example, the second harmonic forms an octave above the fundamental frequency, and the third harmonic forms an interval close to a perfect fifth.

The diagram therefore illustrates how musical harmony arises naturally from the physics of vibrating sound sources. 


```python

import numpy as np
import matplotlib.pyplot as plt

base_freq = 440

harmonic_numbers = np.arange(1,9)
harmonics = base_freq * harmonic_numbers

plt.figure(figsize=(10,5))

plt.stem(harmonic_numbers, harmonics)

plt.xlabel("Harmonic Number")
plt.ylabel("Frequency (Hz)")
plt.title("Harmonic Series and Musical Intervals")

# Label important intervals
plt.text(2, harmonics[1]+50, "Octave\n2:1", ha='center')
plt.text(3, harmonics[2]+50, "Perfect Fifth\n3:2", ha='center')
plt.text(4, harmonics[3]+50, "Perfect Fourth\n4:3", ha='center')

plt.grid(True)
```

fig.2 The diagram shows the harmonic series generated from a fundamental frequency of 440 Hz.Each vertical line in the diagram represents one harmonic. The horizontal axis shows the harmonic number (1st harmonic, 2nd harmonic, etc.), while the vertical axis shows the corresponding frequency in Hertz.
The first harmonic corresponds to the fundamental frequency itself (440 Hz). Each subsequent harmonic is an integer multiple of this base frequency. For example, the second harmonic has frequency 880 Hz, the third harmonic 1320 Hz, and the fourth harmonic 1760 Hz. The third harmonic is Perfect Fifth because the division of it and the previous harmonic would give the ratio 3:2. We could find other Perfect Fifths from lower octaves by finding the lower octave from the third harmonic 1320 Hz / 2 = 660 Hz. Now 660 Hz would be a perfect fifth to 440Hz as well because 660 Hz / 440 Hz would give the 3:2 ratio (equivalent to perfect fifth)

### 3.5 Circle of Fifths

Another important mathematical structure used in music theory is the circle of fifths (Benson, 2007). 
The circle of fifths organizes musical notes according to repeated applications of the perfect fifth interval.

A perfect fifth corresponds to the frequency ratio:

3 : 2

If a note with frequency f is multiplied by this ratio, the resulting note forms a perfect fifth above the original note.

f_new = (3/2)f

By repeatedly applying this relationship, a sequence of notes could be obtained. In the sequence below each successive note is a perfect fifth of the previous.:

C → G → D → A → E → B → F#

This sequence forms the circle of fifths, which plays a central role in tonal harmony and chord progressions. 
Many common harmonic progressions move between chords that are neighbors in the circle of fifths.The circle of fifths therefore provides a mathematical framework for understanding how harmonic relationships are structured in music.


### 3.6 Tonic, Subtonic and Dominant

The tonic (I), subdominant (IV), and dominant (V) chords are closely related through simple mathematical ratios between their frequencies(Benson, 2007). The dominant is a perfect fifth above the tonic (ratio 3:2), and the subdominant is a perfect fourth above the tonic (ratio 4:3) (Benson, 2007; Hall, 2008). These ratios come from the harmonic series and involve small integers, which makes the resulting sounds stable and consonant (Benson, 2007). In this sense, the tonic acts as a base frequency, while the dominant and subdominant are derived from it through simple mathematical relationships. This explains why these three chords form the core of most harmonic progressions in music. In addition there is vi which is considered in music theory as the minor relative to the tonic, and could replace the tonic in order to achieve a more minor harmony.

## 5.Research Hypothesis

This project tests the general notion that musical harmony is governed by simple mathematical relationships between frequencies. Therefore, two tests have been developed in this paper. One for testing whether transitioning into certain intervals within a song is completely random and uniform. In other words, if null hypothesis cannot be rejected, this would mean that there aren't any significant harmonic rules, structures or dependencies. The second one tests whether songs contain random and evenly distributed chord progressions. If the null hypothesis cannot be rejected, in other words this would mean that there aren't any significant harmonic dependencies with chord progressions.


### 5.1. Testing the transition intervals 

Based on the circle of fifths, certain interval transitions are supposed to participate in a given progression more often than others. For example, if a song is in key A, then it would be expected that there will be a transition to its perfect fifth - E or perfect 4th - D.

Therefore the following hypothesis will be tested: 


Null Hypothesis (H₀):
The distribution of chord transition intervals is uniform, meaning that no specific interval occurs more frequently than expected by chance.

Alternative Hypothesis (H₁):
Chord transition intervals are not uniformly distributed, and certain intervals occur significantly more frequently than other intervals.

### 5.2 Testing Chord Progressions

The existence of certain chord progressions is based on the theory of tonic, subdominant and dominant, meaning that certain combinations are expected to exist within a music piece. For example, if the song is in let's say key A (which is the tonic - I), then most probably there will be progression of A (ronic) - D (subdominant) - E (dominant).

Therefore the following hypothesis will be tested:

Null Hypothesis (H₀):
Chord combinations in music are distributed randomly, and there is no preference for specific chord progressions or combinations.

Alternative Hypothesis (H₁):
Chord combinations in music are not random and are significantly dominated by specific chord progressions or combinations.

## 6. Methodology

### 6.1. Statistical Analysis of Intervals

Chord transitions were converted into Roman numeral notation to standardize harmonic relationships across different keys. Consecutive duplicate chords were removed to focus only on meaningful transitions that make a change. Intervals between successive chords were calculated in semitones and analyzed using frequency distributions and probability measures.To determine whether interval transitions occur randomly, a chi-square goodness-of-fit test was applied, comparing observed interval frequencies with a uniform distribution.


### 6.2 Chord Progression Pattern Analysis


To identify common harmonic structures, all combinations of three consecutive chords were extracted from each song. These combinations were compared against all possible triads within the diatonic scale (from I to vii). Their frequencies were analyzed statistically, and a chi-square test was again used to evaluate whether chord combinations follow a random distribution.Additionally, Z-score analysis was used to identify chord progressions that occur significantly more often than average.

## 7. Results


### 7.1 Interval change analysis

The statistical analysis (see Appendix B) of chord transitions reveals a clear pattern in the distribution of intervals. The most frequently occurring intervals are  5, and 7 semitones.
The interval of 7 semitones, corresponding to the perfect fifth, appears with high frequency. Similarly, the interval of 5 semitones, corresponding to the perfect fourth, is also highly represented. These intervals correspond to simple frequency ratios (3:2 and 4:3), which are fundamental in the harmonic series. The prominence of these intervals supports the theoretical expectation that musical harmony is strongly influenced by simple mathematical relationships between frequencies. Smaller intervals such as 2 semitones also appear frequently, representing stepwise harmonic movement. Overall, the results demonstrate that chord progressions are not random but are structured around mathematically significant intervals, particularly the perfect fifth and perfect fourth. The results grom the chi-square analysis show a statistically significant difference between observed and expected frequencies (p < 0.05), leading to rejection of the null hypothesis. This indicates that interval occurrences are not random and not uniform, but instead concentrated around specific values, particularly 5 and 7 semitones.

### 7.2 Chord progression analysis

To evaluate whether chord combinations are randomly distributed, a chi-square test was performed comparing observed frequencies to a uniform distribution (see appendix C). The results from the chi-square test show a significant deviation from randomness (p < 0.05), indicating that certain chord combinations occur much more frequently than expected. Z-score analysis, being over 3, further confirms that combinations such as I–IV–V and I–IV–vi occur significantly more often than average, highlighting their importance in musical harmony. Together, these statistical results support the rejection of the null hypothesis and confirm that musical harmony in terms of chords progressions tends to be more structured rather than random.

Data quality also presents minor limitations, as some entries contained incomplete metadata and were labeled as “Unknown,” which may introduce slight inconsistencies in interpretation. Finally, the statistical approach assumes a uniform distribution as the null hypothesis, which may oversimplify the complexity of musical composition and does not account for stylistic or genre-specific variations. Finally, the methodology presented in chord progression analysis, doesn't extract the exact chord progressions for a given part of the song verse, chorus, etc. but rather extracts the whole song's chord progression in one line. Therefore extracting all groups of three-chord combinations might not be the most exact method. An even more precise method is two extract the chord progressions meant for a specific part of the song.

### Appendix A: Methodology for Interval transitions

```python
import xml.etree.ElementTree as ET
import os
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

dataset_path = "xml"


all_progressions = []

for file in os.listdir(dataset_path):

    if file.endswith(".xml"):

        file_path = os.path.join(dataset_path, file)

        if os.path.getsize(file_path) == 0:
            continue

        try:
            tree = ET.parse(file_path)
            root = tree.getroot()

            chords = root.findall(".//chord")

            progression = []

            for chord in chords:
                sd = chord.find("sd")
                if sd is not None and sd.text:
                    progression.append(sd.text)

            if progression:
                all_progressions.append(progression)

        except ET.ParseError:
            continue




mapping = {
"1":"I", "2":"ii", "3":"iii",
"4":"IV", "5":"V", "6":"vi", "7":"vii°"
}

converted = []

for prog in all_progressions:
    converted.append([mapping[x] for x in prog if x in mapping])



cleaned = []

for prog in converted:
    if not prog:
        continue

    new_prog = [prog[0]]

    for chord in prog[1:]:
        if chord != new_prog[-1]:
            new_prog.append(chord)

    cleaned.append(new_prog)



transitions = []

for prog in cleaned:
    for i in range(len(prog) - 1):
        transitions.append((prog[i], prog[i+1]))

print("Total transitions:", len(transitions))



roman_to_num = {
    "I":0, "ii":2, "iii":4,
    "IV":5, "V":7, "vi":9
}

intervals = []

for a, b in transitions:
    if a in roman_to_num and b in roman_to_num:
        interval = (roman_to_num[b] - roman_to_num[a]) % 12
        intervals.append(interval)



interval_counts = Counter(intervals)
total = sum(interval_counts.values())


labels = sorted(interval_counts.keys())
values = [interval_counts[k] for k in labels]

plt.figure(figsize=(10,5))
plt.bar(labels, values)

plt.xlabel("Interval (Semitones)")
plt.ylabel("Frequency")
plt.title("Chord Transition Interval Frequencies")

plt.show()



probabilities = [interval_counts[k]/total for k in labels]

plt.figure(figsize=(10,5))
plt.bar(labels, probabilities)

plt.xlabel("Interval (Semitones)")
plt.ylabel("Probability")
plt.title("Probability Distribution of Chord Transitions")

plt.show()



print("\nInterval Probabilities:\n")

for k in labels:
    print(f"Interval {k}: {interval_counts[k]/total:.4f}")





probabilities_full = np.array([
    interval_counts.get(i, 0) / total for i in range(12)
])

mean_p = np.mean(probabilities_full)
std_p = np.std(probabilities_full)

print("\nMean probability:", mean_p)
print("Standard deviation of probability:", std_p)



print("\nZ-scores (based on probability):\n")

for i in range(12):
    p = interval_counts.get(i, 0) / total
    z = (p - mean_p) / std_p if std_p != 0 else 0
    print(f"Interval {i}: p = {p:.4f}, Z = {z:.2f}")

from scipy.stats import chisquare


observed = np.array([
    interval_counts.get(i, 0) for i in range(12)
])


expected = np.array([total / 12] * 12)


chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("\nChi-square test results:")
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_value)

print("\nObserved vs Expected:\n")

for i in range(12):
    print(f"Interval {i}: observed = {observed[i]}, expected = {expected[i]:.2f}")
```

### Appendix B: Methodology for chord progressions

```python
from collections import Counter
from itertools import combinations
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare


all_chords = ["I","ii","iii","IV","V","vi","vii°"]


all_possible = list(combinations(all_chords, 3))
all_possible = [tuple(sorted(c)) for c in all_possible]

print("Total possible unique 3-chord progressions:", len(all_possible))  # 35



combo_counts = Counter()

for prog in cleaned:

    unique_chords = set(prog)

    for combo in all_possible:
        if all(chord in unique_chords for chord in combo):
            combo_counts[combo] += 1



total_combos = sum(combo_counts.values())

print("\nTotal observed progressions:", total_combos)



labels = ["-".join(combo) for combo in all_possible]
values = [combo_counts.get(combo, 0) for combo in all_possible]

# Convert to probabilities
probabilities = [v / total_combos if total_combos > 0 else 0 for v in values]



plt.figure(figsize=(14,6))

plt.bar(labels, probabilities)

plt.xticks(rotation=90)
plt.xlabel("3-Chord Combinations")
plt.ylabel("Probability")
plt.title("Distribution of All 3-Chord Combinations (Including vii°)")

plt.tight_layout()
plt.show()



print("\nAll combinations:\n")

for combo, count in zip(all_possible, values):
    p = count / total_combos if total_combos > 0 else 0
    print(f"{'-'.join(combo)}: {count} ({p:.4f})")



observed = np.array(values)

k = len(observed)  # 35
expected = np.array([total_combos / k] * k)

chi2_stat, p_value = chisquare(observed, expected)

print("\nChi-square statistic:", chi2_stat)
print("p-value:", p_value)





print("\nSample observed vs expected:\n")

for combo, o, e in list(zip(all_possible, observed, expected))[:10]:
    print(f"{'-'.join(combo)}: observed={o}, expected={e:.2f}")



mean = np.mean(observed)
std = np.std(observed)

print("\nZ-scores (sample):\n")

for combo, o in list(zip(all_possible, observed))[:10]:
    z = (o - mean) / std if std != 0 else 0
    print(f"{'-'.join(combo)}: Z = {z:.2f}")


```

### Appendix C: Dataset Songs

```python
import xml.etree.ElementTree as ET
import os

dataset_path = "xml"

songs_set = set()


sections = [
    "verse", "chorus", "bridge", "intro", "outro",
    "interlude", "prechorus", "pre-chorus"
]

for file in os.listdir(dataset_path):

    if file.endswith(".xml"):

        file_path = os.path.join(dataset_path, file)

        if os.path.getsize(file_path) == 0:
            continue

        try:
            tree = ET.parse(file_path)
            root = tree.getroot()

            artist = root.find(".//artist")
            title = root.find(".//title")

           
            artist_name = artist.text.strip() if (artist is not None and artist.text) else "Unknown Artist"
            song_title = title.text.strip() if (title is not None and title.text) else "Unknown Title"

  
            title_lower = song_title.lower()

            for sec in sections:
                if sec in title_lower:
                    # remove the section word
                    idx = title_lower.find(sec)
                    song_title = song_title[:idx].strip()
                    break

            # Also clean trailing symbols
            song_title = song_title.replace("-", "").strip()

            formatted = f"{artist_name} – {song_title}"

            songs_set.add(formatted)

        except ET.ParseError:
            continue


songs_list = sorted(songs_set)


for song in songs_list:
    print(song)

print("\nTotal unique songs:", len(songs_list))
```
